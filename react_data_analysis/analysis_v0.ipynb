# US Population - Sex, Race, (Age)

```python
census_data_init = pd.read_csv("2020_decennial_census/DECENNIALPL2020.P2_data_with_overlays_2021-09-16T164642.csv")
census_data_init.iloc[1:].head()
```

```python
just_pop = census_data_init.iloc[1:].filter(items=['NAME', 'P2_001N'])
just_pop = just_pop.rename(columns={'NAME': 'state', 'P2_001N': 'population'})
just_pop.to_csv('census20pop.csv')
just_pop.head()
```

```python
census_filtered = census_data_init.iloc[1:].filter(items=[
    'NAME', 
    'P2_001N', # total pop
    'P2_002N', # hispanic or latino
    # 3 is not-hispanic, 4 is one-race
    'P2_005N', # white
    'P2_006N', # black
    'P2_007N', # indigneous
    'P2_008N', # asian
    'P2_009N', # pacific islander
    'P2_010N', # other
    'P2_011N', # 2+ races
    ])

census_data = census_filtered.rename(columns={
    'NAME': 'state',
    'P2_001N': 'total',
    'P2_002N': 'hispanic',
    'P2_005N': 'white',
    'P2_006N': 'black',
    'P2_007N': 'indigenous',
    'P2_008N': 'asian',
    'P2_009N': 'hawaiian',
    'P2_010N': 'other',
    'P2_011N': 'multiple',
})

census_data = census_data.astype({
    'total': 'int', 
    'hispanic': 'int', 
    'white': 'int', 
    'black': 'int', 
    'indigenous': 'int', 
    'asian': 'int', 
    'hawaiian': 'int', 
    'other':'int', 
    'multiple':'int'})

start_table = census_data.sum()
start_table
```

```python
acs_2019_init = pd.read_csv('ACS_2019/ACSDP1Y2019.DP05_data_with_overlays_2021-08-06T114339.csv').iloc[1:]
acs_2019_init.head()

acs_sex_filtered = acs_2019_init.filter(items=[
    'DP05_0001E', # total pop
    'DP05_0002E', # men
    'DP05_0003E', # women
    ])

acs_sex_perc_filtered = acs_2019_init.filter(items=[
    'DP05_0002PE', # % men
    'DP05_0003PE', # % women
])

# ACS ALSO HAS AGE

acs_race_filtered = acs_2019_init.filter(items=[
    'DP05_0070E', # total pop
    'DP05_0071E', # hispanic or latino
    'DP05_0077E', # white
    'DP05_0078E', # black
    'DP05_0079E', # indigenous
    'DP05_0080E', # asian
    'DP05_0081E', # Hawaiian
    'DP05_0082E', # other
    'DP05_0083E', # multiple races
])
acs_race_perc_filtered = acs_2019_init.filter(items=[
    'DP05_0071PE', # hispanic or latino
    'DP05_0077PE', # white
    'DP05_0078PE', # black
    'DP05_0079PE', # indigenous
    'DP05_0080PE', # asian
    'DP05_0081PE', # Hawaiian
    'DP05_0082PE', # other
    'DP05_0083PE', # multiple races
])
# note: just change to PE for percent
```

```python
acs_sex = acs_sex_filtered.rename(columns={
    'DP05_0001E': 'total', # total pop
    'DP05_0002E': 'num_men', # men
    'DP05_0003E': 'num_women', # women
})

acs_sex_perc = acs_sex_perc_filtered.rename(columns={
    'DP05_0002PE': 'perc_men', # % men
    'DP05_0003PE': 'perc_women', # % women
})

acs_race = acs_race_filtered.rename(columns={
    'DP05_0070E': 'total', # total pop
    'DP05_0071E': 'hispanic', # hispanic or latino
    'DP05_0077E': 'white', # white
    'DP05_0078E': 'black', # black
    'DP05_0079E': 'indigenous', # indigenous
    'DP05_0080E': 'asian', # asian
    'DP05_0081E': 'hawaiian', # Hawaiian
    'DP05_0082E': 'other', # other
    'DP05_0083E': 'multiple', # multiple races
})

acs_race_perc = acs_race_perc_filtered.rename(columns={
    'DP05_0071PE': 'hispanic', # hispanic or latino
    'DP05_0077PE': 'white', # white
    'DP05_0078PE': 'black', # black
    'DP05_0079PE': 'indigenous', # indigenous
    'DP05_0080PE': 'asian', # asian
    'DP05_0081PE': 'hawaiian', # Hawaiian
    'DP05_0082PE': 'other', # other
    'DP05_0083PE': 'multiple', # multiple races
})
```

```python
comp = pd.DataFrame(start_table).drop("state").join(acs_race.transpose())

comp['perc census'] = comp[0] / 3347351.55
comp[1] = comp[1].astype(int)
comp['perc acs'] = comp[1] / 3282395.23
comp.rename(columns={0: 'census2020', 1: 'acs2019'}, inplace=True)
comp

#SIGH. Conclusion: a couple races have shifted by a couple percentage points enough that it's worth using the 2020 data
```

```python
final_sex_perc = acs_sex_perc.iloc[0].astype(float)
final_race_perc = comp['perc census'].drop("total")

sex_json = final_sex_perc.to_json(open("../senate-react/src/data0/sex0.json", 'w'))
race_json = final_race_perc.to_json(open("../senate-react/src/data0/race0.json", 'w'))
```
