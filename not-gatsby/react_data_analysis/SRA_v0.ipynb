```python
import pandas as pd

pop_data = pd.read_csv("census20pop.csv", index_col = 0)

pop_data.head()
```

```python
# this is definitely not how to do this
has_senators = pop_data[pop_data["state"] != "District of Columbia"]
has_senators = has_senators[has_senators["state"] != "Puerto Rico"]

unit = has_senators["population"].sum() / 100
unit
```

```python
has_senators["sra_number"] = has_senators["population"].apply(lambda x: 1 if x / unit < 1 else round(x / unit))
has_senators
```
