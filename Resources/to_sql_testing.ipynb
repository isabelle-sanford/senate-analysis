```python
# imports
import pandas as pd
import sqlalchemy
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# Create Engine for census data
path = "AllData.sqlite"
engine = create_engine(f"sqlite:///{path}")
conn = engine.connect()
```

```python
senate = pd.read_csv('SPopulation1.csv')
attributes = pd.read_csv('CensusAttributesData.csv')
religion = pd.read_csv('ReligiousAffiliationByState.csv')
```

```python
attributes.head()
```

```python
senate.reset_index().to_sql("senate", con = conn, if_exists = 'replace')
attributes.reset_index().to_sql("attributes", con = conn, if_exists = 'replace')
religion.to_sql("religion", con = conn, if_exists = 'replace')

# why does it need to call from the root folder

```
